## <center> DFT BÁSICO </center> 
## <u>Ejercicio 1</u>:
Dada la señal discreta, $x(t)$ de un 1 segundo de duración, compuesta por tres funciones $x1(t)+x2(t)+x3(t)$ que se reescriben a continuación, y muestreada a Fs = 44.1 kHz:

$x_1 (t) = 2$ \
$x_2 (t) = cos(2 \pi f_2 t) e^{- \frac{(t-\mu_2)^2}{2\sigma_2^2}}\;\;\;\;\;\;\;\;f_2 = 10kHz$ $\mu_2 = 0.2s$ $\sigma_2 = 0.05s$ \
$x_3 (t) = sen(2 \pi f_3 t) e^{- \frac{(t-\mu_3)^2}{2\sigma_3^2}}\;\;\;\;\;\;\;\;f_3 = 10.1kHz$ $\mu_3 = 0.7s$ $\sigma_3 = 0.07s$

Graficar la magnitud (en escala lineal y en dB) y la fase de la transformada de Fourier (DFT) de la señal total. Indicar qué representan las componentes no nulas. El eje de frecuencia debe estar en Hz.

## <u>Respuesta</u>: 

### Generacion de la senal x(t):
Se comienza generando la señal $x(t)$ como pide el ejercicio, con una duración de un segundo y una frecuencia de muestreo de 44100 Hz.

```python
# Duracion y frecuencia de muestreo
time = 1
fs = 44100
vector_t = np.linspace(0,time,time*fs)

# Parametros de cada senal
f2 = 10000
mu2 = 0.2
sigma2 = 0.05

f3 = 10100
mu3 = 0.7
sigma3 = 0.07

exp2 = ((vector_t-mu2)**2)/(2*(sigma2**2))
exp3 = ((vector_t-mu3)**2)/(2*(sigma3**2))

x1 = 2*np.ones_like(vector_t)
x2 = np.cos(2*np.pi*f2*vector_t)*(np.e**(-exp2))
x3 = np.sin(2*np.pi*f3*vector_t)*(np.e**(-exp3))
x = x1+x2+x3
```

Luego, utilizando la librería Numpy, se utiliza la función 'rfft'. La FFT es un algoritmo muy eficiente que permite calcular la DFT de forma rapida, siendo su resultado identico al de la DFT. Por lo tanto, se utiliza la FFT a lo largo de este trabajo. <br>
Cabe aclarar que esta función tiene como salida la mitad derecha del resultado de la fft, es decir, solo las frecuencias positivas. Ya que para el análisis que se busca, no son necesarias las frecuencias negativas y no tienen representación práctica en la vida real. 

```python
# Implementacion de Scipy de la FFT
fft_x = np.fft.rfft(x)
freqs_x = np.fft.rfftfreq(time*fs, 1/fs)

# En dB
valuedB = 20*np.log10(abs(fft_x))
```

### Plot de la DFT en escala lineal, dB y su fase:

```python
# Plot de la DFT
plt.rcParams.update({'font.size': 22})
fig, axs = plt.subplots(3, 1, sharex=True, figsize=[20, 15])
axs[0].plot(freqs_x, np.abs(fft_x))
axs[0].set_title('DFT de $x(t)$')
axs[1].plot(freqs_x, valuedB, color='red')
axs[1].set_title('DFT de $x(t)$ en dB')
axs[2].plot(freqs_x, np.angle(fft_x), color='green')
axs[2].set_title('Fase de la DFT de $x(t)$')

fig.tight_layout()

axs.flat[0].set(ylabel='Amplitud')
axs.flat[1].set(ylabel='Amplitud [dB]')
axs.flat[2].set(ylabel='Fase [rad]', xlabel='Frecuencia [Hz]', yticks=(-np.pi,np.pi))

for ax in axs.flat:
    ax.grid()
    
plt.show()
```

### Conclusión:
La componente en cero representa la señal continua $x_1$ de frecuencia "0 Hz" (por ser continua constante). <br>
Por otro lado, en la DFT de escala lineal, se puede apreciar las componentes de frecuencia 10 y 10.1 kHz de las señales $x_2$ y $x_3$ respectivamente. <br>
Sin embargo, en la DFT de escala en dB se puede ver fácilmente el leaking generado en la transformada, que se encuentra cercano a las frecuencias de 10 y 10.1 kHz y tiene forma de lóbulo. Una forma de interpretar ese lóbulo es como si estuvieramos convolucionando la respuesta en frecuencia de una ventana cuadrada (similar a una funcion sinc con un lóbulo principal centrado en cero) con dos deltas de dirac en 10 y 10.1 kHz respectivamente.

## <center> INTERPOLACIÓN CUADRÁTICA </center>
## <u>Ejercicio 3</u>:
Ahora, nos quedaremos con la componente $x_3(t)$ del inciso anterior para implementar una interpolación que mejore la resolución en frecuencia del máximo de la DFT. La señal será entonces de 1 segundo de duración y muestreada a $Fs = 44.1 kHz$:<br><br>

<center>$x_3 (t) = sen(2 \pi f_3 t) e^{- \frac{(t-\mu_3)^2}{2\sigma_3^2}}\;\;\;\;\;\;\;\;f_3 = 10008.37Hz$&emsp;$\mu_3 = 0.7s$&emsp;$\sigma_3 = 0.07s$ 


Se deberá graficar la magnitud, expresada en dB, de la DFT de la señal $x_3(t$) y se determinará la frecuencia a la que se detecta el máximo de la magnitud, verificándose que no es la frecuencia exacta. En casos como este, en donde la resolución no es suficiente para discriminar el valor de una componente de frecuencia, se puede mejorar la determinación de la posición mediante técnicas de interpolación. Una de las más simples, pero efectivas, es utilizar la interpolación cuadrática que se calcula con la siguiente ecuación: <br><br>

<center> $\hat{n}_{max} = n_{max} + \frac{\alpha - \gamma}{2(\alpha - 2\beta + \gamma)}$

En donde,\
nmax : posición encontrada del punto máximo.\
$\beta = g[nmax]$ valor absoluto de la magnitud en dB en la posición máxima nmax del pico. \
$\alpha = g[nmax-1]$ valor absoluto de la magnitud en dB en la posición anterior al máximo del pico. \
$\gamma = g[nmax+1]$ valor absoluto de la magnitud en dB en la posición posterior al máximo del pico.


**Nota:** esta ecuación está definida para un intervalo unitario (muestreo o resolución). Si dicho valor no es 1, debe multiplicarse el valor correspondiente al segundo sumando de la ecuación.

## <u>Respuesta</u>: 
Se genera la señal a partir de la copia de la  señal del ejercicio anterior.

### Generacion de la senal y su DFT en dB:
En el gráfico de esta sección, se puede observar claramente que no se puede distinguir una frecuencia de la otra. Para ilustrar aun más este concepto, se realizan dos figuras, una con la señal entera y una con un foco en la frecuencia de 10 kHz.

```python
# Generacion de senal
x_ej3 = x3_ej2

# DFT y conversion a dB
ej3_dft_x = np.fft.rfft(x_ej3)
ej3_freqs = np.fft.rfftfreq(len(x_ej3), 1/fs)
ej3_dft_x = 20 * np.log10(np.maximum(np.abs(ej3_dft_x), 1e-10))
```

```python
# Plot
plt.rcParams.update({'font.size': 22})
fig, axs = plt.subplots(2, 1, figsize=[15, 10])
axs[0].plot(ej3_freqs, ej3_dft_x, color='orange')
axs[1].plot(ej3_freqs[9980:10040], ej3_dft_x[9980:10040], color='red')

axs[0].set_title('DFT de $x_3(t)$')
axs[1].set_title('Zoom en las frecuencias de interes de la DFT de $x_3(t)$')

axs[0].set(xlabel='Frecuencia[Hz]', ylabel='Amplitud [dB]')
axs[1].set(xlabel='Frecuencia[Hz]', ylabel='Amplitud [dB]')

axs[0].grid()
axs[1].grid()

fig.tight_layout()
plt.show()
```

### Cálculo de la Interpolación Cuadrática:
En base a la ecuación dada en la consigna, se definen los tres parámetros. Para encontrar el valor de *nmax* (posición del punto máximo) se utiliza la funcion *np.argmax* de la libreria Numpy que precisamente devuelve el valor del índice para el cual se encuentra el máximo de la señal. Luego, este índice se utiliza en la DFT para definir los valores de beta, gamma y alfa. Finalmente se calcula el valor de *nmax* aproximado y se lo imprime en la salida de la consola.

```python
# Busca la frecuencia maxima y defino los parametro beta, alfa y gamma.
nmax = np.argmax(ej3_dft_x)
beta = ej3_dft_x[nmax]
alfa = ej3_dft_x[nmax-1]
gamma = ej3_dft_x[nmax+1]

# Ecuacion de la interpolacion
nmax_aprox = nmax + ((alfa-gamma)/(2*(alfa - 2*beta + gamma)))
print('El resultado de nmax aproximado es de: ', round(nmax_aprox,3), 'Hz')
```

### Conclusión:
El resultado muestra que la interpolación acerca el valor desde el entero 10009 Hz a su valor original. Se puede ver también que la aproximación no es perfecta ya que no tiene los mismos valores decimales. A terminos prácticos, se tiene un valor más preciso de la frecuencia real utilizada.

## <center>CONVOLUCIÓN MEDIANTE DFT</center>
## <u>Ejercicio 5</u>:

Calcular la convolución lineal del punto 7 del TP1, pero esta vez utilizando el algoritmo de la DFT. Determinar el largo que deben tener las señales a convolucionar para que el resultado de utilizar una convolución circular vía DFT sea equivalente a la convolución lineal.

## <u>Respuesta</u>: 
Para rearmar la señal utilizada en el punto 7 del TP1, se utiliza $x_1$ y $x_2$ de este TP (ya que son las mismas) y se crea $x_3$ del TP1 para agregarla.

```python
# Senal del ejercicio 7 del TP1
ej5_f3 = 500
ej5_u3 = 0.7
ej5_sigma3 = 0.07
ej5_exp3 = ((vector_t-ej5_u3)**2)/(2*(ej5_sigma3**2))

ej5_x3 = np.sin(2*np.pi*f3*vector_t)*(np.e**(-exp3))
ej5_x = x1+x2+ej5_x3

# Generacion de Ventana rectangular de largo M
ej5_M = 9
ej5_rectangular = np.ones(ej5_M)/ej5_M
```

El largo que debe tener la señal para ser una convolución lineal está determinado por el largo de ambas señales.
En este caso, llamaremos $L$ al largo de la senal $x(t)$ y $M$ al largo de la ventana.
Para que este algoritmo se comporte como una convolución lineal, se debe cumplir que:<br><br> 

<center>$N = M + L -1$</center>

Siendo $N$ la longitud de la convolución. A continuación, se realizaran los cálculos con el agregado de ceros (zero padding).

Por otro lado, la convolución por DFT parte de las propiedades matemáticas que tiene la transformada de Fourier, quedando de esta forma:<br><br>

<center>$\mathscr{F}\{x(t)*h(t)\} = X[n]H[n]$</center>


```python
# Completo x[n] y h[n] con ceros hasta llegar a N=M+L-1.
N = ej5_M + len(ej5_x) - 1
ej5_x = np.pad(ej5_x, (0, int(N-len(ej5_x))))
ej5_rectangular = np.pad(ej5_rectangular, (0, int(N-len(ej5_rectangular))))

# Calculo la DFT de cada senal
ej5_dft_rectangular = np.fft.fft(ej5_rectangular)
ej5_dft_x = np.fft.fft(ej5_x)

# Multiplicar X[k] con H[k] para obtener Y[k]
ej5_dft_conv = ej5_dft_x*ej5_dft_rectangular

# Realizo la IDFT para obtener y[n]
ej5_conv = np.real(np.fft.ifft(ej5_dft_conv))
```

Cabe aclarar que se toma el valor real de la inversa de la DFT. Idealmente, este resultado no tendría valores complejos, pero debido a errores de aproximación por trabajar con señales discretas, tiene valores imaginarios pequeños que deben descartarse.

```python
# Plot
plt.rcParams.update({'font.size': 22})
fig, axs = plt.subplots(3, 1, figsize=[15, 15])
axs[0].plot(vector_t, ej5_x[:-ej5_M+1], color='orange')
axs[1].plot(np.arange(len(ej5_conv))/fs, ej5_conv, color='red')
axs[2].plot((np.arange(len(ej5_conv))/fs)[ej5_M:-ej5_M], ej5_conv[ej5_M:-ej5_M], color='violet')
axs[0].set_title('$x(t)$ sin zero padding')
axs[1].set_title('Convolucion lineal con la FFT de $x(t)$')
axs[2].set_title('Convolucion lineal recortada de $x(t)$')

for i in range(3):
    axs[i].set(xlabel='Tiempo [s]', ylabel='Amplitud')
    axs[i].grid()

fig.tight_layout()
plt.show()
```

### Conclusión:
Se puede notar que la convolución es definitivamente lineal en el segundo gráfico debido a los valores que se encuentran en el borde de la señal. Los métodos numéricos para resolver ecuaciones diferenciales deben poder calcular las derivadas mediante algún método aproximado. Uno de los más métodos más conocidos es el de diferencias finitas, que si bien es simple, no siempre tiene la precisión deseada. Otro método, es el de la diferenciación en frecuencia que hace uso de las propiedades de la transformada de Fourier. En este punto deberán tomar la señal $x_2(t)$ de un 1 segundo de duración y muestreada a $Fs=44.1 kHz$:

$x_2 (t) = cos(2 \pi f_2 t) e^{- \frac{(t-\mu_2)^2}{2\sigma_2^2}}\;\;\;\;\;\;\;\;f_2 = 10kHz$&emsp;$\mu_2 = 0.2s$&emsp;$\sigma_2 = 0.05s$ 

y calcular la derivada analítica, la primera diferencia hacia adelante (forward) y la derivada por el
método de la transformada de Fourier. Graficar los resultados y compara la aproximación de la

## <u>Respuesta</u>: 
Se copia la señal del ejercicio 2, luego se utiliza la definición de la derivada analítica para ese tipo de señales que se provee con la consigna del trabajo como ayuda. Se guarda la variable con el nombre *dx_dt*. <br>
El segundo método y el tercer método se implementan con a partir de la referencia [4] utilizando Numpy.

```python
# Defino la senal
ej6_x = np.copy(x2)

# 1er metodo: Derivada analitica
a = 2*np.pi*10000
t = np.linspace(0,1,fs)
b = 0.2
c = 2*(0.05**2)
exp = np.e**(-((t-b)**2)/c)
dx_dt = -a*np.sin(a*t)*exp - (2*(t-b)*np.cos(a*t)*exp)/c

# 2do metodo: diferencia hacia adelante
dx_fd = np.empty(44100)
for i in range(len(ej6_x)-1):
    dx_fd[i] = (ej6_x[i+1]-ej6_x[i]) / (1/fs)
dx_fd[-1] = dx_fd[-2]

# 3er metodo: metodo por la DFT
ej6_fft_len = fs
fhat = np.fft.fft(ej6_x)
kappa = (2*np.pi)*np.arange(-fs/2,(fs/2))
kappa = np.fft.fftshift(kappa)
dfhat = (1j * kappa * fhat)
dx_fft = np.real(np.fft.ifft(dfhat))
```

```python
plt.rcParams.update({'font.size': 22}) # Tamano de fuente 
fig, ax = plt.subplots(3, 1, figsize=[20,15])# Instancia de objeto Figure y Axes.
ax[0].plot(vector_t, dx_dt, color='red')
ax[1].plot(vector_t, dx_fd, color='orange')
ax[2].plot(vector_t, dx_fft, color='black')

ax[0].set_title('Derivada analitica a partir de la definicion')
ax[1].set_title('Derivada a partir de la diferencia hacia adelante')
ax[2].set_title('Derivada por el Metodo por la DFT')

for i in range(3):
    ax[i].set(xlabel='Tiempo [s]', ylabel='Amplitud')
    ax[i].grid(True)

fig.tight_layout()
```

## <center>TRANSFORMADA EN TIEMPO CORTO DE FOURIER (STFT) - ESPECTROGRAMA</center>
## <u>Ejercicio 7</u>:
A la señal del inciso 1 de este TP calcular y graficar la magnitud en dB y la fase de su STFT. Realizar tres gráficos cambiando el tipo de ventana y el largo del cálculo de la FFT. Determinar alguna combinación de las tres anteriores que sea (parezca) mejor para representar a la señal en estudio.

## <u>Respuesta</u>: 

En primer lugar, se copia la señal del ejercicio 1. Luego, se generan las STFT correspondientes. <br>
Para esto se utiliza la funcion STFT de la libreria Scipy, que dentro de sus parametros permite elegir la ventana utilizada para recortar cada porción de la señal y la longitud de cada FFT que se calcula para lograr la STFT. <br>
En este caso particular se elige la ventana rectangular y la ventana de blackman para comparar y dos longitudes fijadas de 256 y 4410. Siendo 256 la longitud por defecto de la función y 4410 un 10% de la longitud de la senal total.

```python
ej7_x = np.copy(x)

# Con ventana Rectangular y L = 256
ej7_fft_len1 = 256
ej7_freqs1, ejt_time1, Zxx1 = stft(ej7_x, fs, window='boxcar', nperseg=ej7_fft_len1)

# Con ventana Rectangular y L = 4410
ej7_fft_len2 = 4410
ej7_freqs2, ejt_time2, Zxx2 = stft(ej7_x, fs, window='boxcar', nperseg=ej7_fft_len2)

# Con ventana Blackman y L = 4410
ej7_fft_len3 = 4410
ej7_freqs3, ejt_time3, Zxx3 = stft(ej7_x, fs, window='blackman', nperseg=ej7_fft_len3)
```

Para graficar las transformadas, una vez mas se utiliza la libreria matplotlib y sus subplots para comparar mejor cada senal. En este caso se utiliza la funcion 'pcolormesh' que grafica un mapa de color con tres ejes: Frecuencia y Tiempo (x,y) y un eje z representado por el color que sera de Amplitud en dB o Fase en radianes depende de lo que corresponda.

```python
plt.rcParams.update({'font.size': 18}) # Tamano de fuente 
fig, ax = plt.subplots(3, 2, figsize=[20,20]) # Instancia de objeto Figure y Axes.

# Grafico mapa de color para cada STFT
pcm00 = ax[0,0].pcolormesh(ejt_time1, ej7_freqs1, 20*np.log10(np.abs(Zxx1)), shading='gouraud', cmap='viridis')
pcm01 = ax[0,1].pcolormesh(ejt_time1, ej7_freqs1, np.angle(Zxx1), shading='gouraud', cmap='RdBu_r')
pcm10 = ax[1,0].pcolormesh(ejt_time2, ej7_freqs2, 20*np.log10(np.abs(Zxx2)), shading='gouraud', cmap='viridis')
pcm11 = ax[1,1].pcolormesh(ejt_time2, ej7_freqs2, np.angle(Zxx2), shading='gouraud', cmap='RdBu_r')
pcm20 = ax[2,0].pcolormesh(ejt_time3, ej7_freqs3, 20*np.log10(np.abs(Zxx3)), shading='gouraud', cmap='viridis')
pcm21 = ax[2,1].pcolormesh(ejt_time3, ej7_freqs3, np.angle(Zxx3), shading='gouraud', cmap='RdBu_r')

# Titulo de cada mapa
ax[0,0].set_title('STFT con ventana rectangular y L = '+str(ej7_fft_len1))
ax[0,1].set_title('Fase de la STFT con Ventana rectangular y L = '+str(ej7_fft_len1))
ax[1,0].set_title('STFT con ventana rectangular y L = '+str(ej7_fft_len2))
ax[1,1].set_title('Fase de la STFT con Ventana Rectangular y L = '+str(ej7_fft_len2))
ax[2,0].set_title('STFT con ventana Blackman y L = '+str(ej7_fft_len3))
ax[2,1].set_title('Fase de la STFT con Blackman y L = '+str(ej7_fft_len3))

# Nombre de ejes x,y
for i in range(2):
    ax[i,0].set(xlabel='Tiempo [s]', ylabel='Frecuencia [Hz]')
    ax[i,1].set(xlabel='Tiempo [s]', ylabel='Frecuencia [Hz]')

# Barra de color con su unidad
fig.colorbar(pcm00, ax=ax[0,0], label='Amplitud [dB]')
fig.colorbar(pcm01, ax=ax[0,1], label='Fase [rad]')
fig.colorbar(pcm10, ax=ax[1,0], label='Amplitud [dB]')
fig.colorbar(pcm11, ax=ax[1,1], label='Fase [rad]')
fig.colorbar(pcm20, ax=ax[2,0], label='Amplitud [dB]')
fig.colorbar(pcm21, ax=ax[2,1], label='Fase [rad]')
     
fig.tight_layout()
plt.show()
```

### Conclusión:
En el caso de dejar muy grande la longitud de cada ventana para la cual se calcula la FFT (como se ve en la figura 3 con L=4410), se nota que en los bordes de la STFT, la transformada pierde informacion. Sin embargo, a mayor la longitud de laventana, permite ver mas finamente la frecuencia. 

La eleccion de L (largo de la fft), dependera del contexto a utilizar. Si se necesita una resolucion mas clara en la frecuencia se debe utilizar una ventana mas grande, pero si se necesitan los valores en los bordes, se debe negociar con una ventana no tan grande.

En el caso de la eleccion de la ventana, al elegir la ventana rectangular en el principio y el final de la transformada se obtienen valores mas representativos de la senal. Sin embargo, la rectangular tiene peor definicion en la amplitud de las frecuencias. Esto se debe a que despues del lobulo principal, los lobulos secundarios se encuentran a mayor amplitud que la ventana de blackman (viendo la DFT de la ventana rectangular). Esto puede interpretarse como que el leaking en este Colormesh afecta mas al resultado visual de este ejercicio en particular.
